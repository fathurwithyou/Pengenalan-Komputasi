# Problem 3

## Import Required Modules

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('H05.csv')
```

### No 1

```python
df['nilai_mat'].plot(kind='hist',
                     title='Distribusi Nilai Matematika',
                     bins=[i for i in range(0, 101, 5)],
                    edgecolor='black')
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.show()
```

### No 2

```python
df['fakultas'].value_counts().plot(
    kind='barh',
    title='Diagram Batang Horizontal Jumlah Mahasiswa setiap Fakultas',
    edgecolor='black')
plt.show()
```

### No 3

```python
df['fakultas'].value_counts().plot(
    kind='pie', title='Diagram Pie Jumlah Mahasiswa setiap Fakultas',wedgeprops={'linewidth': 2, 'edgecolor': 'black'})
plt.show()
```

### No 4

Berdasarkan gambar di atas, **FTI merupakan fakultas dengan jumlah mahasiswa terbanyak**. Menurut saya, lebih baik menggunakan diagram batang horizontal karena kita ingin mengetahui jumlah tiap-tiap fakultas sehingga kita dapat secara "effortless" untuk mengetahui fakultas mana yang memiliki mahasiswa terbanyak.

### No 5

```python
df.plot(kind='scatter',
        x='nilai_kim',
        y='nilai_fis',
        title='Scatter Plot Nilai Kimia dengan Nilai Fisika')
plt.show()
```
