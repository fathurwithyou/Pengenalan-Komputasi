```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('P05-03.csv')
df.head()
```

## No 1

```python
data = df[df['room_type'] == 'Private room']
data['number_of_reviews'].plot(kind='hist',
                               x='number_of_reviews',
                               bins=[i for i in range(0, 300, 30)],
                               edgecolor='black',
                               color='orange')
plt.show()
```

## No 2

```python
df[df['number_of_reviews'] > 100].loc[:, ['room_type']].value_counts().plot(
    kind='pie', wedgeprops={
        "linewidth": 1,
        "edgecolor": "black"
    }, autopct='%.1f%%')
plt.title('Perbandingan banyaknya masing-masing jenis penyediaan ruang dengan banyaknya review di aplikasi di atas 100')
plt.show()
```

## No 3

```python
data = df[df['host_name'] != 'Jay']
data.plot(kind='scatter',
          x='longitude',
          y='latitude',
          edgecolor='black',
          color='orange')
plt.show()
```

## No 4

```python
data = df[df['neighbourhood'] == 'Novena']
data.plot(kind='scatter',
          x='longitude',
          y='latitude',
          sizes=data['availability_365'],
          color='orange',
          edgecolor='black')
plt.show()
```

Idenya adalah mengondisikan kolom dengan hanya berisi *Novena*, kemudian kita membuat bubble plot dengan ketersediaan rumah selama satu tahun sebagai ukuran dari bubble.

## No 5

```python
data = df[(df['host_name'] == 'Jojo') | (df['host_name'] == 'Elizabeth') |
          (df['host_name'] == 'Jimmy')]
data.groupby(['room_type', 'host_name']).size().unstack().plot(kind='bar',
                                                               stacked=True)
plt.show()
```
